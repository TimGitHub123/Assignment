### My code print graphic of the function y = sin(x)\*sinh(x)

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.arange(-10, 10, 0.01, float)
y = np.sin(x)*np.sinh(x)
plt.plot(x, y)
plt.show()
```
